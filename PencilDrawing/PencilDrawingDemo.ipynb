```python
from PencilDrawingBySketchAndTone import *
import matplotlib.pyplot as plt
```

```python
jl_img = io.imread('./PencilDrawing/images/jleauge.jpg')
av_img = io.imread('./PencilDrawing/images/avengers.jpg')
dp_img = io.imread('./PencilDrawing/images/deadpool.jpg')
```

```python
pencil_tex = './PencilDrawing/pencils/pencil3.jpg'
jl_im_pen = gen_pencil_drawing(jl_img, kernel_size=8, stroke_width=1, num_of_directions=8, smooth_kernel="gauss",
                       gradient_method=1, rgb=True, w_group=2, pencil_texture_path=pencil_tex,
                               stroke_darkness=2, tone_darkness=1.5)
# jl_im_pen[jl_im_pen > 1] = 1
# jl_im_pen[jl_im_pen < -1] = -1
jl_im_pen = exposure.rescale_intensity(jl_im_pen, in_range=(0,1))
plt.rcParams['figure.figsize'] = [16,12]
plt.subplot(1,2,1)
plt.imshow(jl_img)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(jl_im_pen)
plt.axis("off")
plt.show()
```

```python
pencil_tex = './PencilDrawing/pencils/pencil0.jpg'
av_im_pen = gen_pencil_drawing(av_img, kernel_size=8, stroke_width=2, num_of_directions=8, smooth_kernel="gauss",
                       gradient_method=1, rgb=True, w_group=2, pencil_texture_path=pencil_tex,
                               stroke_darkness=2, tone_darkness=1.5)
av_im_pen[av_im_pen > 1] = 1
av_im_pen[av_im_pen < -1] = -1
plt.rcParams['figure.figsize'] = [16,12]
plt.subplot(1,2,1)
plt.imshow(av_img)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(av_im_pen)
plt.axis("off")
plt.show()
```

```python
pencil_tex = './PencilDrawing/pencils/pencil3.jpg'
dp_im_pen = gen_pencil_drawing(dp_img, kernel_size=8, stroke_width=0, num_of_directions=8, smooth_kernel="gauss",
                       gradient_method=1, rgb=True, w_group=0, pencil_texture_path=pencil_tex,
                               stroke_darkness=2, tone_darkness=1.5)
dp_im_pen[dp_im_pen > 1] = 1
dp_im_pen[dp_im_pen < -1] = -1
plt.rcParams['figure.figsize'] = [16,12]
plt.subplot(1,2,1)
plt.imshow(dp_img)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(dp_im_pen)
plt.axis("off")
plt.show()
```
